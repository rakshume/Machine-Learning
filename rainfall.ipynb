```python
import matplotlib.pyplot as plt
import pandas as pd
import random 
```

```python
# Generate random rainfall data for each value in 'a'
rainfall = [random.uniform(0, 100) for _ in a]

# Store the data in a dataframe
df = pd.DataFrame({'Day': a, 'Rainfall': rainfall})
df.tail()
```

```python
plt.plot(df['Day'], df['Rainfall'], marker='o')
plt.xlabel('Day')
plt.ylabel('Rainfall')
plt.title('Rainfall Over Days')
plt.grid(True)
plt.show()
```
