```python
# Load .mat file (Replace 'your_file.mat' with actual filename)
data = scipy.io.loadmat('c:\Users\POOJA\.cache\kagglehub\datasets\patrickfleith\nasa-battery-dataset\versions\2\cleaned_dataset')

# Extract relevant data (Modify based on actual dataset keys)
re_z = data['ReZ'].flatten()  # Real part of Impedance
im_z = data['ImZ'].flatten()  # Imaginary part of Impedance
cycle_count = data['CycleCount'].flatten()  # Aging cycle count

# Create 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(re_z, im_z, cycle_count, c=cycle_count, cmap='jet', marker='o')

# Labels and title
ax.set_xlabel("Re(Z) - Real Impedance (Ohm)")
ax.set_ylabel("Im(Z) - Imaginary Impedance (Ohm)")
ax.set_zlabel("Cycle Count (Aging)")
ax.set_title("3D Plot of Impedance vs Aging")

plt.show()
```
